```python
def get_temp_cat(temp):
    if temp < 0:
        return 'Freezing'
    elif temp < 15:
        return 'Chilly'
    elif temp < 26:
        return 'Nice'
    elif temp >= 26:
        return 'Hot'
    else:
        return temp
# создаем функцию для перекодировки температуры в качественную ее оценку
```

```python
bikes['Temperature Category'] = bikes['Temperature'].apply(get_temp_cat)
bikes[['Temperature Category', 'Temperature']].tail()
```

```python
q1 = bikes['Rental Count'].quantile(0.25)
q3 = bikes['Rental Count'].quantile(0.75)
iqr = q3 - q1
```

```python
iqr_outlier_threshold_bottom = q1 - 1.5 * iqr
iqr_outlier_threshold_up = q3 + 1.5 * iqr
print(iqr_outlier_threshold_bottom)
print(iqr_outlier_threshold_up)

# определение выбросов через полтора межквартильных размаха, выбросов вниз нет, т.к. там граница меньше нуля
```

```python
bikes[bikes['Rental Count'] > iqr_outlier_threshold_up].shape

# определяем сколько строк являются выбросами
```

```python
mean = bikes['Rental Count'].mean()
std = bikes['Rental Count'].std()
std_outlier_threshold_bottom = mean - 2 * std
std_outlier_threshold_up = mean + 2 * std
print(std_outlier_threshold_bottom)
print(std_outlier_threshold_up)

# определение выбросов через стандартное отклонение (сигмы), где 2-3 - умеренный выброс, больше трех - жесткий
# запомнить, что сигмы откладываются от среднего, а не от медианы
```

```python
iqr_bikes_no_outliers = bikes[bikes['Rental Count'] <= iqr_outlier_threshold_up]
std_bikes_no_outliers = bikes[bikes['Rental Count'] <= std_outlier_threshold_up]
```

```python
print(bikes['Rental Count'].mean())
print(iqr_bikes_no_outliers['Rental Count'].mean())
print(std_bikes_no_outliers['Rental Count'].mean())
print(bikes['Rental Count'].median())
print(iqr_bikes_no_outliers['Rental Count'].median())
print(std_bikes_no_outliers['Rental Count'].median())

# смотрим как отличаются медиана и среднее для массивов с выбросами и без
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
# библиотека сиборн основана на нампае и работает на ней, нужна для более крутой визуализации 
import matplotlib.pyplot as plt
```

```python
N = 10 ** 4
np.random.seed(42)

def plot_comparison(x, title):
    fig, ax = plt.subplots(3, 1, sharex=True, figsize=(15,10))
    sns.histplot(x=x, ax=ax[0], kde=True)
    # гистограмма - histplot против обычного hist в мэтплотлиб. КДЕ добавляет скользящее среднее
    # ax = ax[0] это обращение к нужной строке в описании фигуры, т.е. к нужному графику
    ax[0].set_title('Histogram + KDE')
    sns.boxplot(x=x, ax=ax[1])
    # боксплот - ящик с усами
    ax[1].set_title('Boxplot')
    # сет тайтл нужна для создания заголовка
    sns.violinplot(x=x, ax=ax[2])
    # скрипичный график
    ax[2].set_title('Violin plot')
    fig.suptitle(title, fontsize=16)
    plt.show()
    
sample_gaussian = np.random.normal(size=N)
plot_comparison(sample_gaussian, 'Standard Normal Distribution')
```
